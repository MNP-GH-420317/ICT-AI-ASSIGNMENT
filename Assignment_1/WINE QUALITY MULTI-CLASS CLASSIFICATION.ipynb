```python
# import modules 
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import sklearn
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression,LogisticRegression
from sklearn.tree import DecisionTreeClassifier,DecisionTreeRegressor
from sklearn.datasets import load_wine
from sklearn.metrics import classification_report,accuracy_score,root_mean_squared_error,confusion_matrix
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import GridSearchCV
```

```python
# dataset will arrange 
wine = load_wine()
data = pd.DataFrame(wine.data,columns=wine.feature_names)
data.head(3)
```

```python
# To add Qality in DataSet 
data['quality'] = wine.target
```

```python
# Logistic Regression, Decision Trees, or K-Nearest Neighbors
```

```python
model_normal = {
    'LogisticRegression' : LogisticRegression(),
    'DecisionTreeClassifier': DecisionTreeClassifier(),
    'KNeighborsClassifier' : KNeighborsClassifier()
}

x =  data.drop('quality',axis=1)
y = data['quality']

data_stand = StandardScaler()

x_std_scalar = data_stand.fit_transform(x)

x_train,x_test , y_train , y_test = train_test_split(x_std_scalar,y,test_size=0.8,random_state=16)

model_summary_normal = {}

for model_name , model_ in model_normal.items():
    model_fit =  model_.fit(x_train,y_train)
    y_prd_nor = model_fit.predict(x_test)
    accuracy =  accuracy_score(y_prd_nor,y_test)
    model_summary_normal[model_name] = accuracy
    print(f'-------> {model_name} <---------')
    print(classification_report(y_prd_nor,y_test))
```

```python
model_summary_normal
```

```python
# model confusion metrix will using 
sns.heatmap(confusion_matrix(y_prd_nor,y_test),annot=True)
plt.title('confusion_matrix')
plt.show()
```

```python
# using dataset Gridseacrch cv 
model_summary_grd = {}

models = {
    'LogisticRegression' : LogisticRegression(),
    'DecisionTreeClassifier': DecisionTreeClassifier(),
    'KNeighborsClassifier' : KNeighborsClassifier()
}

model_para = {
    "LogisticRegression": {
        'C': [0.1, 1, 10],
        'solver': ['lbfgs', 'liblinear']
    },
    "DecisionTreeClassifier": {
        'criterion': ['gini', 'entropy', 'log_loss'],
        'splitter': ['best', 'random'],
        'max_depth': [2, 10, 15, None],
        'min_samples_split': [2, 3, 4, 5],
        'max_features': ['sqrt', 'log2', None]
    },
    "KNeighborsClassifier": {
        'n_neighbors': [3, 4, 5, 6],
        'weights': ['uniform', 'distance'],
        'algorithm': ['ball_tree', 'kd_tree', 'brute'],
        'leaf_size': [10, 20, 30]
    }
}

for model_name , model in models.items():
    model_grid = GridSearchCV(
        estimator=model,
        param_grid = model_para[model_name],
        cv=5,
        scoring='accuracy',
        n_jobs= -1)
    ml = model_grid.fit(x_train,y_train)
    y_prd_grd = ml.predict(x_test)
    print( f"{model_name} <-> {model_grid.best_estimator_} ")
    # print( f"{model_name} <-> {model_grid.best_params_} ")
    accuracy_grd =  accuracy_score(y_prd_nor,y_test)
    model_summary_grd[model_name] = accuracy_grd
```

```python
model_summary_grd
```

```python
# model confusion metrix will using 
sns.heatmap(confusion_matrix(y_prd_grd,y_test),annot=True)
plt.title('confusion_matrix')
plt.show()
```
